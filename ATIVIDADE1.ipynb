**19) Crie duas listas e apresente em uma nova lista:**

**a. A soma os elementos da lista**

**21) Crie na forma de polinômio usando o comando `poly1d` os seguintes polinômios:**

**a. 3x + 4**

**b. 4x^3 + 3x^2 + 2x + 1**

**c. 2x^3 + 3**

**d. x^2 + 2x + 3**

**e. Multiplique os polinômios c e d**

**f. Derive o polinômio d**

**g. Integre o polinômio d**

```python
import numpy as np

# a. 3x + 4
poli_a = np.poly1d([3, 4])
print("\na. Polinômio a:", poli_a)

# b. 4x^3 + 3x^2 + 2x + 1
poli_b = np.poly1d([4, 3, 2, 1])
print("\nb. Polinômio b:", poli_b)

# c. 2x^3 + 3
poli_c = np.poly1d([2, 0, 0, 3])
print("\nc. Polinômio c:", poli_c)

# d. x^2 + 2x + 3
poli_d = np.poly1d([1, 2, 3])
print("\nd. Polinômio d:", poli_d)

# e. Multiplique os polinômios c e d
poli_e = poli_c * poli_d
print("\ne. Multiplicação de c e d:", poli_e)

# f. Derive o polinômio d
derivada_d = poli_d.deriv()
print("\nf. Derivada de d:", derivada_d)

# g. Integre o polinômio d
integral_d = poli_d.integ()
print("\ng. Integral de d:", integral_d)
```

**8) Apresente os gráficos das funções de Besel e da função de erro Gaussiana**

```python
from scipy.special import jv, erf
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = jv(0, x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('J0(x)')
plt.title('Função de Bessel de ordem 0')
plt.grid(True)
plt.show()

x = np.linspace(-3, 3, 100)
y = erf(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('erf(x)')
plt.title('Função de erro Gaussiana')
plt.grid(True)
plt.show()
```

**10) Apresente os gráficos dos polinômios de Legendre**

```python
from scipy.special import legendre

x = np.linspace(-1, 1, 100)

for n in range(5):
  Pn = legendre(n)
  y = Pn(x)
  plt.plot(x, y, label=f"P{n}(x)")

plt.xlabel('x')
plt.ylabel('Pn(x)')
plt.title('Polinômios de Legendre')
plt.grid(True)
plt.legend()
plt.show()
```

**11) Apresente os gráficos das funções PDF e CDF (CDF é a integral de PDF (Função de Densidade de Probabilidade) e PDF é a derivada de CDF).**

```python
from scipy.stats import norm

x = np.linspace(-3, 3, 100)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Função Densidade de Probabilidade (PDF) - Distribuição Normal')
plt.grid(True)
plt.show()

plt.plot(x, cdf)
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Função de Distribuição Acumulada (CDF) - Distribuição Normal')
plt.grid(True)
plt.show()
```

**12) Apresente o gráfico de probabilidade da função de massa de probabilidade (PMF) para a distribuição binomial (Probabilidade x Número de Sucessos).**

```python
from scipy.stats import binom

n = 10
p = 0.5

x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.plot(x, pmf, 'bo-', label='PMF Binomial')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.title('Gráfico de Probabilidade da Distribuição Binomial')
plt.xticks(x)
plt.grid(True)
plt.legend()
plt.show()
```

**14) Apresente os valores de correlação de Pearson e Spearman e um gráfico com linha de regressão ajustada sobre os pontos de dados.**

**a. A correlação de Pearson avalia a relação linear entre duas variáveis contínuas.**

**b. A correlação de Spearman avalia a relação monotônica entre duas variáveis contínuas ou ordinais.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.5

correlacao_pearson, p_valor_pearson = pearsonr(x, y)
print("Correlação de Pearson:", correlacao_pearson)

correlacao_spearman, p_valor_spearman = spearmanr(x, y)
print("Correlação de Spearman:", correlacao_spearman)

coeficientes = np.polyfit(x, y, 1)
linha_regressao = np.poly1d(coeficientes)

plt.scatter(x, y, label='Pontos de dados')
plt.plot(x, linha_regressao(x), color='red', label='Linha de regressão')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.title('Gráfico com Linha de Regressão')
plt.legend()
plt.show()
```

1) MLP: Usar (Carregar e Importar) a base de dados parkinsons

```python
# Caminho para o arquivo
caminho_arquivo = "parkinsons.data"
df_parkinsons = pd.read_csv(caminho_arquivo, sep=",")

# Exibir as 5 primeiras linhas
df_parkinsons.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 2. Normalizar todas as colunas entre 0 e 1
scaler = MinMaxScaler()
df_parkinsons_normalized = pd.DataFrame(scaler.fit_transform(df_parkinsons.iloc[:, 1:]), columns=df_parkinsons.columns[1:])
df_parkinsons_normalized["status"] = df_parkinsons["status"]
df_parkinsons_normalized
```

```python
# 3. Separar X (features) e y (target)
X = df_parkinsons_normalized.drop(columns=["status"])
y = df_parkinsons_normalized["status"]
```

```python
# 4. Dividir o conjunto em treinamento e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Conjunto de treinamento: {X_train.shape}")
print(f"Conjunto de teste: {X_test.shape}")
```

```python
from sklearn.neural_network import MLPClassifier

# 5. Configurar a topologia da Rede Neural e parâmetros
mlp = MLPClassifier(
    hidden_layer_sizes=(100,),  # Topologia: 1 camada escondida com 100 neurônios
    activation="relu",          # Função de ativação: ReLU
    solver="adam",              # Otimizador: Adam
    max_iter=500,               # Iterações máximas
    random_state=42
)
```

```python
# 6. Treinar o modelo
mlp.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 7. Fazer predições e calcular acurácia
y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

print(f"Percentual de acerto: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico 1: Matriz de Confusão
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Saudável", "Parkinson"], yticklabels=["Saudável", "Parkinson"])
plt.title("Matriz de Confusão")
plt.ylabel("Verdadeiro")
plt.xlabel("Predito")
plt.show()
```

```python
# Gráfico 2: Erro durante o treinamento
plt.figure(figsize=(6, 5))
plt.plot(mlp.loss_curve_)
plt.title("Curva de Perda Durante o Treinamento")
plt.xlabel("Iterações")
plt.ylabel("Erro")
plt.show()
```

```python
# Gráfico 3: Comparação de predições
plt.figure(figsize=(6, 5))
sns.countplot(x=y_test, label="Verdadeiro", color="blue", alpha=0.6)
sns.countplot(x=y_pred, label="Predito", color="orange", alpha=0.6)
plt.title("Comparação de Verdadeiros x Preditos")
plt.legend()
plt.show()
```

2) MLP: Usar (Carregar e Importar) a base de dados penguin (API seaborn)

```python
# Caminho para o arquivo
caminho_arquivo = "penguins.csv"
df_penguins = pd.read_csv(caminho_arquivo, sep=",")

# Exibir as 5 primeiras linhas
df_penguins.head()
```

```python
df_penguins.info()
```

```python
df_penguins.isnull().sum()
```

```python
df_penguins.dropna(inplace=True)
df_penguins.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# 3. Tratamento de dados categóricos
# Transformar colunas categóricas em numéricas usando LabelEncoder
label_encoders = {}
for col in ["species", "island", "sex"]:
    le = LabelEncoder()
    df_penguins[col] = le.fit_transform(df_penguins[col])
    label_encoders[col] = le
```

```python
# 4. Normalizar todas as colunas numéricas entre 0 e 1
scaler = MinMaxScaler()
penguins_scaled = pd.DataFrame(scaler.fit_transform(df_penguins.iloc[:, 1:]), columns=df_penguins.columns[1:])
penguins_scaled["species"] = df_penguins["species"]
```

```python
if penguins_scaled["species"].isnull().sum() > 0:
    print("Encontrado valores nulos em species!")
    penguins_scaled = penguins_scaled.dropna(subset=["species"])
```

```python
# 5. Separar X (features) e y (target)
X = penguins_scaled.drop(columns=["species"])
y = penguins_scaled["species"]
```

```python
# 6. Dividir os dados em treinamento e teste (70% treinamento, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Conjunto de treinamento: {X_train.shape}")
print(f"Conjunto de teste: {X_test.shape}")
```

```python
from sklearn.neural_network import MLPClassifier

# 7. Configurar a topologia da Rede Neural
mlp = MLPClassifier(
    hidden_layer_sizes=(50, 30),  # Topologia: 2 camadas (50 e 30 neurônios)
    activation="relu",           # Função de ativação: ReLU
    solver="adam",               # Otimizador: Adam
    max_iter=1000,               # Iterações máximas
    random_state=42
)
```

```python
# 8. Treinar o modelo
mlp.fit(X_train, y_train)
```

```python
# 9. Fazer predições e calcular acurácia
y_pred = mlp.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))
print(f"Percentual de acerto: {accuracy * 100:.2f}%")
```

```python
# Gráfico 1: Matriz de Confusão
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoders["species"].classes_, yticklabels=label_encoders["species"].classes_)
plt.title("Matriz de Confusão")
plt.ylabel("Verdadeiro")
plt.xlabel("Predito")
plt.show()
```

```python
# Gráfico 2: Curva de Perda
plt.figure(figsize=(6, 5))
plt.plot(mlp.loss_curve_)
plt.title("Curva de Perda Durante o Treinamento")
plt.xlabel("Iterações")
plt.ylabel("Erro")
plt.show()
```

```python
# Gráfico 3: Comparação de Predições
plt.figure(figsize=(6, 5))
sns.countplot(x=y_test, label="Verdadeiro", color="blue", alpha=0.6)
sns.countplot(x=y_pred, label="Predito", color="orange", alpha=0.6)
plt.title("Comparação de Verdadeiros x Preditos")
plt.legend()
plt.show()
```
